## Шаг 4

Отделяем кусок данных от столбца school до столбца guardian. Можно просто перечислить столбики, но это же скучно)

Транспонирование может быть тяжёлой операцией, тогда способ так себе, но за то читается легко)

```python
df = df.transpose().iloc[0:12].transpose()
df.head()
```

## Шаг 5-8

Применяем лямбду к столбикам.
apply работает не inplace, а возвращает новый объект (в данном случае series).
Если нужно записать данные в таблицу, то можно просто приравнять соответствующий столбец.

```python
df['Mjob'] = df['Mjob'].apply(lambda x: x.upper())
df['Fjob'] = df['Fjob'].apply(lambda x: x.upper())
```

```python
df.tail()
```

## Шаг 9

Выясняем процентное соотношение по возрастам.
Группируем по возрасту, потом считаем количество записей в каждой группе, потом считаем проценты.
Выбор столбика до начала вычислений должен сократить количество операций)

p.s. Запись в переменную result тут для того, чтобы поменять название столбика

```python
result = df[['age', 'sex']].groupby('age').count()/len(df) * 100
result.columns = ['Проценты']
result
```

## Шаг 10

Поля Walc и Dalc выкинуты на 4 шаге, поэтому сначала снова считаем файл, а затем найдём средне потребление алкоголя.


```python
df = pd.read_csv('student-mat.csv')
```

```python
df[['age', 'Walc', 'Dalc']].groupby('age').mean()
```

## Шаг 11
